## ================== Series & DataFrame Basics ==================

### Dataframe

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [24, 30, 22],
    "City": ["Delhi", "Mumbai", "Bangalore"]
}
print(data)
print("\n-------------------------")
dtdf = pd.DataFrame(data)
print(dtdf)
dtdf
```

### Read Data Set

```python
dataset = pd.read_csv("analysis-dataset.csv")
print(dataset[0:2])
print("\n-------------------------")
df = pd.DataFrame(dataset)
print(df[0:2])
df
```

### Series

```python

row = ["A", "B", "C"]
print(row)
print(type(row))

print("\n-------------------------")
srs = pd.Series(row, name="names")
print(srs)
print(type(srs))

print("\n-------------------------")
sdf = pd.DataFrame(srs)
print(sdf)
print(type(sdf))

sdf
```

```python
df_csv = pd.read_csv("analysis-dataset.csv", header=1)
# Useful if your file had extra headers and you want to skip them
df_csv
```

```python
df_csv = pd.read_csv("analysis-dataset.csv", skiprows=1)
print(df_csv)
# ==> First skip 1 row (col_names) and then from remaining data of table choose the 3rd index (i.e. 4th postioned) row as the header
df_csv = pd.read_csv("analysis-dataset.csv", header = 3, skiprows=1)
df_csv
```

### DICTIONARY

### TUPLES LIST

```python
weather_data = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]
df = pd.DataFrame(data=weather_data, columns=['day','temperature','windspeed','event'])
df
```
